### Working with date

```python
date.info()
```

```python
# Since our dataset, 'date' column is an object so we need to convert it in date-time datatype
```

```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
# now it has been changed to datetime64[ns]
```

#### Extract year

```python
date['date_year'] = date['date'].dt.year
```

#### Extract month

```python
date['date_month_no'] = date['date'].dt.month
```

```python
date['date_month'] = date['date'].dt.month_name()
```

```python
date.head()
```

#### Extract days

```python
date['date_day'] = date['date'].dt.day
```

```python
# dow = day of week
date['date_dow'] = date['date'].dt.dayofweek
date['date_dow_name'] = date['date'].dt.day_name()
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

```python
# checking if that date is weekend or not
date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1, 0)
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

#### Extract quarter 

```python
date['quarter'] = date['date'].dt.quarter
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

#### Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1, 2]), '1st sem', '2nd sem')
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```
